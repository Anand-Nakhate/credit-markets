## b. Add a function to construct generic fixed rate cashflow schedules from symbology data

Use one row of the symbology dataframe as input  to the function. Use the helper function to convert a date string to a QuantLib date object.

```python
def get_ql_date(date) -> ql.Date:
    """
    convert dt.date to ql.Date
    """
    if isinstance(date, dt.date):
        return ql.Date(date.day, date.month, date.year)
    elif isinstance(date, str):
        date = dt.datetime.strptime(date, "%Y-%m-%d").date()
        return ql.Date(date.day, date.month, date.year)
    else:
        raise ValueError(f"to_qldate, {type(date)}, {date}")
```

```python
def create_schedule_from_symbology(details: dict):
    '''Create a QuantLib cashflow schedule from symbology details dictionary (usually one row of the symbology dataframe)
    '''
    # Create maturity from details['maturity']
    maturity = get_ql_date(details['maturity'])    # use date helper function and update code !!!
    
    # Create acc_first from details['acc_first']
    acc_first = get_ql_date(details['acc_first'])   # use date helper function and update code !!!
    
    # Create calendar for Corp and Govt asset classes
    calendar = ql.UnitedStates(ql.UnitedStates.GovernmentBond)
    
    # define period from details['cpn_freq'] ... can be hard-coded to 2 = semi-annual frequency
    if details.get('cpn_freq') == 1:
        period = ql.Period(ql.Annual)
    elif details.get('cpn_freq') == 2:
        period = ql.Period(ql.Semiannual)
    elif details.get('cpn_freq') == 4:
        period = ql.Period(ql.Quarterly)
    else:
        # Default to semiannual if missing or unrecognized
        period = ql.Period(ql.Semiannual)

        
    # business_day_convention
    business_day_convention = ql.Unadjusted
    
    # termination_date_convention
    termination_date_convention = ql.Unadjusted
    
    # date_generation
    date_generation=ql.DateGeneration.Backward
    
    # Create schedule using ql.MakeSchedule interface (with keyword arguments)
    schedule = ql.MakeSchedule(effectiveDate=acc_first,  # this may not be the same as the bond's start date
                            terminationDate=maturity,
                            tenor=period,
                            calendar=calendar,
                            convention=business_day_convention,
                            terminalDateConvention=termination_date_convention,
                            rule=date_generation,
                            endOfMonth=True,
                            firstDate=ql.Date(),
                            nextToLastDate=ql.Date())
    return schedule
```

## c. Add a function to construct generic fixed rate bond objects from symbology data

Implement the  create_schedule_from_symbology() function below. 

Identify and display the `AAPL 2.2 09/11/29` bond details in the symbology dataframe. Use create_schedule_from_symbology() to generate and display the cashflow schedule for `AAPL 2.2 09/11/29`.


```python
def create_bond_from_symbology(details: dict):
    '''Create a US fixed rate bond object from symbology details dictionary (usually one row of the symbology dataframe)
    '''
    
     # Create day_count from details['dcc']
     # For US Treasuries use ql.ActualActual(ql.ActualActual.ISMA)
     # For US Corporate bonds use ql.Thirty360(ql.Thirty360.USA)
    dcc = details['dcc'].strip().upper()
    if dcc in ['ACT/ACT']:
        day_count = ql.ActualActual(ql.ActualActual.ISMA)
    elif dcc in ['30/360']:
        day_count = ql.Thirty360(ql.Thirty360.USA)
    else:
        # Default to Thirty360 if the dcc is unrecognized
        day_count = ql.Thirty360(ql.Thirty360.USA)  # update code depending on dcc !!!
    
    # Create day_count from details['start_date']    
    issue_date = get_ql_date(details['start_date'])     # use date helper function and update code !!!
    
    # Create days_settle from details['days_settle']
    days_settle = int(details['days_settle'])        # update code depending on days_settle !!!

    # Create days_settle from details['coupon']
    coupon = float(details['coupon']) / 100.0       # update code depending on coupon !!!

    # Create cashflow schedule
    schedule = create_schedule_from_symbology(details)
    
    face_value = 100
    redemption = 100
    
    payment_convention = ql.Unadjusted
        
    # Create fixed rate bond object
    fixed_rate_bond = ql.FixedRateBond(
        days_settle,
        face_value,
        schedule,
        [coupon],
        day_count,
        payment_convention,
        redemption,
        issue_date)        

    return fixed_rate_bond
```

## d. Add function that returns a dataframe with (future) cash flows details for a bond object

Use the "Investigate Bond Cashflows" section in the Quantlib introductory notebook as a template.

The results dataframe should contain following columns:


| CashFlowDate | CashFlowAmount | CashFlowYearFrac |
|----------|-------|-------------|

Create the bond objects and display the future cashflows for following securities:

* the `T 5 3/8 02/15/31` government bond and 
* the `AAPL 2.2 09/11/29` corporate bond.

```python
def get_bond_cashflows(bond: ql.FixedRateBond, calc_date: ql.Date) -> pd.DataFrame:
    """
    Returns all future cashflows as of calc_date, i.e. with payment dates > calc_date.
    The returned DataFrame has columns:
      - CashFlowDate (string in YYYY-MM-DD format)
      - CashFlowAmount (float)
      - CashFlowYearFrac (float, approximate years from calc_date)
    """

    # Set the global evaluation date (optional but common in QuantLib usage)
    ql.Settings.instance().evaluationDate = calc_date

    cf_dates = []
    cf_amounts = []
    cf_year_fracs = []

    for cf in bond.cashflows():
        pay_date = cf.date()
        if pay_date > calc_date:
            # Collect the date and amount
            cf_dates.append(pay_date)
            cf_amounts.append(cf.amount())

            # Approximate the year fraction (day count) from calc_date
            day_diff = pay_date.serialNumber() - calc_date.serialNumber()
            year_frac = day_diff / 365.0
            cf_year_fracs.append(year_frac)

    # Build the DataFrame
    data = {
        'CashFlowDate': [d.ISO() for d in cf_dates],  # convert QuantLib dates to YYYY-MM-DD strings
        'CashFlowAmount': cf_amounts,
        'CashFlowYearFrac': cf_year_fracs
    }
    cashflows_df = pd.DataFrame(data)
    return cashflows_df
```

## b. Calibrate the on-the-run treasury yield curve (bootstrapping)
The function below shows how to calibrate a smooth yield/discount factor curve from the on-the-run treasury dataframe. 

Calibrate the bid, ask and mid discount factor curves as of 2024-12-13.

Display the calibration results for the mid curve, using get_yield_curve_details_df().

```python
def calibrate_yield_curve_from_frame(
        calc_date: ql.Date,
        treasury_details: pd.DataFrame,
        price_quote_column: str):
    '''Create a calibrated yield curve from a details dataframe which includes bid/ask/mid price quotes.
    '''
    ql.Settings.instance().evaluationDate = calc_date

    # Sort dataframe by maturity
    sorted_details_frame = treasury_details.sort_values(by='maturity')    
    
    # For US Treasuries use ql.ActualActual(ql.ActualActual.ISMA)
    day_count = ql.ActualActual(ql.ActualActual.ISMA)

    bond_helpers = []
    
    for index, row in sorted_details_frame.iterrows():
        bond_object = create_bond_from_symbology(row)
        
        tsy_clean_price_quote = row[price_quote_column]
        tsy_clean_price_handle = ql.QuoteHandle(ql.SimpleQuote(tsy_clean_price_quote))
        
        bond_helper = ql.BondHelper(tsy_clean_price_handle, bond_object)
        bond_helpers.append(bond_helper)
        
    yield_curve = ql.PiecewiseLogCubicDiscount(calc_date, bond_helpers, day_count)
    # yield_curve = ql.PiecewiseFlatForward(calc_date, bond_helpers, day_count)
    
    yield_curve.enableExtrapolation()
    return yield_curve


def get_yield_curve_details_df(yield_curve, curve_dates=None):
    
    if(curve_dates == None):
        curve_dates = yield_curve.dates()

    dates = [d.to_date() for d in curve_dates]
    discounts = [round(yield_curve.discount(d), 3) for d in curve_dates]
    yearfracs = [round(yield_curve.timeFromReference(d), 3) for d in curve_dates]
    zeroRates = [round(yield_curve.zeroRate(d, yield_curve.dayCounter(), ql.Compounded).rate() * 100, 3) for d in curve_dates]

    yield_curve_details_df = pd.DataFrame(data={'Date': dates,
                             'YearFrac': yearfracs,
                             'DiscountFactor': discounts,
                             'ZeroRate': zeroRates})                             
    return yield_curve_details_df
```

```python
# 1. Select the calibration date and filter the DataFrame for that date
calc_date = ql.Date(13, 12, 2024)
selected_date_str = "2024-12-13"

treasury_data_for_date = merged_on_the_run_treasury_prices[
    merged_on_the_run_treasury_prices['date'] == selected_date_str
]

# 2. Calibrate yield curves using bid, ask, and mid prices
yield_curve_bid = calibrate_yield_curve_from_frame(calc_date, treasury_data_for_date, "bidPrice")
yield_curve_ask = calibrate_yield_curve_from_frame(calc_date, treasury_data_for_date, "askPrice")
yield_curve_mid = calibrate_yield_curve_from_frame(calc_date, treasury_data_for_date, "midPrice")

# 3. Display calibration results for the mid curve
yield_curve_details_mid = get_yield_curve_details_df(yield_curve_mid)
print("Mid Curve Calibration Results:")
print(yield_curve_details_mid)
```

## d. Plot calibrated discount factors

Plot the discount factor curve up to the 30 years point, using a 6 months discretization grid.

```python
# 1. Define end date ~30 years from calc_date
end_date = calc_date + ql.Period(30, ql.Years)

# 2. Create a 6-month interval grid of dates
dates = []
discounts = []
year_fracs = []

current_date = calc_date
while current_date <= end_date:
    dates.append(current_date)
    # Convert current_date to time in years from reference
    t = yield_curve_mid.timeFromReference(current_date)
    year_fracs.append(t)
    
    # Retrieve the discount factor at this date
    dcf = yield_curve_mid.discount(current_date)
    discounts.append(dcf)
    
    # Increment by 6 months
    current_date = current_date + ql.Period(6, ql.Months)

# 3. Build a DataFrame (optional, for clarity)
df = pd.DataFrame({
    'Date': [d.ISO() for d in dates],
    'YearFrac': year_fracs,
    'DiscountFactor': discounts
})

# 4. Plot the discount factors vs. time in years
plt.figure(figsize=(10, 6))
plt.plot(df['YearFrac'], df['DiscountFactor'], marker='o', linestyle='-')
plt.xlabel('Time-to-Maturity (Years)')
plt.ylabel('Discount Factor')
plt.title('Calibrated Mid Curve Discount Factors (Up to 30 Years)')
plt.grid(True)
plt.show()

```

```python
# For US Treasuries, use ActualActual(ISMA) day count
day_count = ql.ActualActual(ql.ActualActual.ISMA)
compounding = ql.Compounded
coupon_freq = 2  # assuming semiannual coupons

# Prepare lists to store calculated metrics
duration_list = []
convexity_list = []
dv01_list = []

# Loop over each treasury bond row in the dataframe
for index, row in treasury_pricing.iterrows():
    # Create the bond object from symbology row (a Pandas Series)
    bond_obj = create_bond_from_symbology(row)
    bond_obj.setPricingEngine(bond_engine)
    
    # Calculate the clean price using the pricing engine
    clean_price = bond_obj.cleanPrice()
    
    # Compute yield to maturity (YTM) from the clean price
    ytm = ql.BondFunctions.bondYield(bond_obj, clean_price, day_count, compounding, coupon_freq)
    
    # Create an InterestRate object from the computed YTM
    ytm_rate = ql.InterestRate(ytm, day_count, compounding, coupon_freq)
    
    # Calculate analytical duration and convexity using the computed YTM rate
    bond_duration = ql.BondFunctions.duration(bond_obj, ytm_rate)
    bond_convexity = ql.BondFunctions.convexity(bond_obj, ytm_rate)
    
    # Compute DV01 as Dirty Price * Duration
    dirty_price = bond_obj.dirtyPrice()
    dv01 = dirty_price * bond_duration
    
    duration_list.append(bond_duration)
    convexity_list.append(bond_convexity)
    dv01_list.append(dv01)

# Extend the treasury_pricing DataFrame with the calculated risk metrics
treasury_pricing['duration'] = duration_list
treasury_pricing['convexity'] = convexity_list
treasury_pricing['dv01'] = dv01_list

# Display the new columns alongside market data for comparison
print(treasury_pricing[['midPrice', 'calc_mid_price', 'duration', 'convexity', 'dv01']])
```

| scen_dv01 | scen_duration | scen_convexity |

```python
def compute_scenario_sensitivities(
    bond: ql.FixedRateBond,
    base_curve_handle: ql.YieldTermStructureHandle,
    bump_size: float = 1e-4  # 1 bp in decimal
):
    """
    Computes scenario-based DV01, Duration, and Convexity for a bond
    using +/- bump_size parallel shifts on the base_curve_handle.
    Returns (dv01, duration, convexity).
    """

    # 1) Create a Quote for the parallel shift
    interest_rate_bump = ql.SimpleQuote(0.0)

    # 2) Build a zero-spreaded term structure that can be shifted
    spreaded_curve = ql.ZeroSpreadedTermStructure(
        base_curve_handle,
        ql.QuoteHandle(interest_rate_bump)
    )
    scenario_engine = ql.DiscountingBondEngine(
        ql.YieldTermStructureHandle(spreaded_curve)
    )

    # 3) Attach the scenario engine to the bond
    bond.setPricingEngine(scenario_engine)

    # --- Base Price (no shift) ---
    interest_rate_bump.setValue(0.0)
    price_base = bond.cleanPrice()
    dirty_price_base = bond.dirtyPrice()

    # --- Price in +1bp scenario ---
    interest_rate_bump.setValue(bump_size)
    price_up = bond.cleanPrice()

    # --- Price in -1bp scenario ---
    interest_rate_bump.setValue(-bump_size)
    price_down = bond.cleanPrice()

    # Reset the bump to zero (optional)
    interest_rate_bump.setValue(0.0)

    # DV01: from the snippet => (price_down_1bp - price_base) * 1e4 / 100
    # Here we store it as a positive measure of sensitivity.
    dv01 = (price_down - price_base) * 1e4 / 100.0

    # Duration = DV01 / dirty_price_base * 100
    duration = (dv01 / dirty_price_base) * 100.0

    # Convexity: from the snippet => gamma_1bp = (price_down - 2*price_base + price_up) * 1e8 / 100
    gamma_1bp = (price_down - 2.0*price_base + price_up) * 1e8 / 100.0
    convexity = (gamma_1bp / dirty_price_base) * 100.0

    return dv01, duration, convexity
```

```python
# Prepare lists to store scenario metrics
scen_dv01_list = []
scen_duration_list = []
scen_convexity_list = []

for idx, row in treasury_pricing.iterrows():
    # Create the bond object
    bond_obj = create_bond_from_symbology(row)
    
    # Compute scenario sensitivities
    dv01, duration, convexity = compute_scenario_sensitivities(
        bond_obj,
        ql.YieldTermStructureHandle(yield_curve_mid)  # your calibrated mid curve
    )
    
    scen_dv01_list.append(dv01)
    scen_duration_list.append(duration)
    scen_convexity_list.append(convexity)

# Add the scenario results to the DataFrame
treasury_pricing['scen_dv01'] = scen_dv01_list
treasury_pricing['scen_duration'] = scen_duration_list
treasury_pricing['scen_convexity'] = scen_convexity_list

# Display the updated DataFrame with scenario columns
treasury_pricing[['ticker', 'isin', 'midPrice', 'calc_mid_price',
                        'scen_dv01', 'scen_duration', 'scen_convexity']]
```

## b. Compute analytical Yields and Z-Spreads

Compute analytical Yields and Z-Spreads metrics, as described in Section 7. "Analytical Duration, Convexity and Z-Spread (flat yield model)" in the QuantLib Basic notebook.

```python
merged_corp_bonds = pd.merge(
    bond_market_prices_eod,
    corp_bond_symbology, 
    on=['ticker', 'isin', 'figi'],
    how='inner'
)
merged_corp_bonds.columns
```

```python
import QuantLib as ql
import pandas as pd

# For US Corporate bonds, use the Thirty360(USA) day count convention.
day_count = ql.Thirty360(ql.Thirty360.USA)
compounding = ql.Compounded
coupon_frequency = ql.Semiannual  # Use Frequency enum instead of integer

# Set the evaluation date (adjust as needed)
calc_date = ql.Date(6, 4, 2025)
ql.Settings.instance().evaluationDate = calc_date

# Prepare lists to store calculated metrics.
calc_yield_list = []
calc_zspread_list = []

# Loop over each corporate bond in merged_corp_bonds.
for idx, row in merged_corp_bonds.iterrows():
    # Create the bond object from the symbology row (a Pandas Series).
    bond_obj = create_bond_from_symbology(row)
    
    # Attach a pricing engine.
    # Use a temporary flat curve built on an arbitrary rate (e.g., 5.5%) to get the market price.
    temp_flat_rate = 5.5e-2  
    temp_flat_curve = ql.FlatForward(calc_date, temp_flat_rate, day_count, compounding, coupon_frequency)
    temp_flat_curve_handle = ql.YieldTermStructureHandle(temp_flat_curve)
    bond_engine = ql.DiscountingBondEngine(temp_flat_curve_handle)
    bond_obj.setPricingEngine(bond_engine)
    
    # Retrieve the bond's clean market price.
    market_price = bond_obj.cleanPrice()
    
    # Compute the implied yield via price-to-yield conversion.
    implied_yield = bond_obj.bondYield(market_price, bond_obj.dayCounter(), compounding, coupon_frequency)
    calc_yield_list.append(implied_yield * 100)  # Convert to percentage.
    
    # Build a flat yield curve using the computed implied yield.
    computed_flat_curve = ql.FlatForward(calc_date, implied_yield, day_count, compounding, coupon_frequency)
    computed_flat_curve_handle = ql.YieldTermStructureHandle(computed_flat_curve)
    
    # Compute the Z-spread relative to the computed flat yield curve.
    # We pass the settlement date (calc_date) explicitly along with accuracy, max iterations, and a guess.
    zspread = ql.BondFunctions.zSpread(
        bond_obj,
        market_price,
        computed_flat_curve_handle,
        bond_obj.dayCounter(),
        compounding,
        coupon_frequency,
        calc_date,   # settlement date
        1.0e-10,     # accuracy
        100,         # maximum iterations
        0.0          # initial guess
    )
    # Convert z-spread to basis points.
    calc_zspread_list.append(zspread * 10000)

# Extend the merged_corp_bonds DataFrame with the new calculated risk metrics.
merged_corp_bonds['calc_yield'] = calc_yield_list
merged_corp_bonds['calc_zspread'] = calc_zspread_list

# Display a subset of the results for comparison.
print(merged_corp_bonds[['midPrice', 'calc_yield', 'calc_zspread']])
```

## c. Validate Z-Spread computation for a few fixed rate corporate bonds

For the following corporate bonds:

* `IBM 1.95 05/15/30`
* `ORCL 2 7/8 03/25/31`
* `DIS 6.65 11/15/37`

use function below to re-price them using the calibrated flat z-spread. Follow the example in Section 7. "Analytical Duration, Convexity and Z-Spread (flat yield model)".

Validate/confirm that you match the original market prices, which were used as input to the z-Spread function.


```python
def calc_clean_price_with_zspread(fixed_rate_bond, yield_curve_handle, zspread):
    zspread_quote = ql.SimpleQuote(zspread)
    zspread_quote_handle = ql.QuoteHandle(zspread_quote)
    yield_curve_bumped = ql.ZeroSpreadedTermStructure(yield_curve_handle, zspread_quote_handle, ql.Compounded, ql.Semiannual)
    yield_curve_bumped_handle = ql.YieldTermStructureHandle(yield_curve_bumped)
    
    # Set Valuation engine
    bond_engine = ql.DiscountingBondEngine(yield_curve_bumped_handle)
    fixed_rate_bond.setPricingEngine(bond_engine)
    bond_clean_price = fixed_rate_bond.cleanPrice()
    return bond_clean_price
```
